```python
########################### Adding parameters to class ####################
# --- First parameter in any method is 'self'
class Phone:
    def make_call(self):
        print("Let's make a phone call")
    def play_game(self):
         print("Let's play a game")
    def set_color(self, color):
        self.color = color
    def set_cost(self, cost):
        self.cost = cost
    def show_color(self):
        print(self.color)
    def show_cost(self):
        print(self.cost)
```

```python
# Object instanstiation
p2 = Phone()
```

```python
p2.set_color("blue")
p2.set_cost(521)
p2.play_game()
p2.set_color("red")
p2.set_cost(800)
p2.show_color()
p2.show_cost()
```

```python
# ------------------- CREATING CLASS WITH CONSTRUCTORS------------
class Employee:
    def __init__(self, name,age, salary, gender):
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
    def employee_details(self):
        print("Name of the employee is ",self.name)
        print("Age of ", self.name, "is ", self.age)
        print("Salary of ",self.name, "is ", self.salary)
        print("Gender of ", self.name,  "is ",self.gender)
```

```python
e1 = Employee("Karina",25,11200,"Female")
e1.employee_details()
```

```python
# -------------Inheritance--------------

class Vehicle:
    # Constructor
    def __init__(self, milage, cost):
        self.milage = milage
        self.cost = cost
    # methods to show the details of vehicle
    def show_details(self):
        print("This is a vehicle")
        print("The milage is", self.milage)
        print("The cost is ",self.cost)
```

```python
v1 = Vehicle(300,54000)
v1.show_details()
```

```python
# Creating a child class of Vehicle
class Car(Vehicle):
    def show_car(self):
        print("This is a car")
```

```python
# Instantiation of object of child class
c1 = Car(250,29500)
c1.show_car() # Invoking the child class method
c1.show_details() # Invoking parent class method
```

```python
# Over-riding init method
class Truck(Vehicle):
    def __init__(self, mileage, cost,tyre,hp):
        super().__init__(mileage,cost)
        self.tyre = tyre
        self.hp = hp
    def show_truck(self):
        print("This is a truck")
        print("The number of typers are", self.tyre)
        print("The horse power is", self.hp)
```

```python
t1 = Truck(500,50000,8,2000)
t1.show_details()
```

```python
t1.show_truck()
```
