```python
############### OOPs #########################
# Creating a first class"Phone" in python - class name should be in camel case
class Phone:
    def make_call(self):
        print("Making phone call")

    def play_game(self):
        print("Playing game")
```

```python
# object instantiation of class
p1 = Phone()
```

```python
# invoking methods of class
p1.make_call()
```

```python
p1.play_game()
```

```python
########################### Adding parameters to class ####################
# --- First parameter in any method is 'self'
class Phone:
    def make_call(self):
        print("Let's make a phone call")
    def play_game(self):
         print("Let's play a game")
    def set_color(self, color):
        self.color = color
    def set_cost(self, cost):
        self.cost = cost
    def show_color(self):
        print(self.color)
    def show_cost(self):
        print(self.cost)
```

```python
# Object instanstiation
p2 = Phone()
```

```python
p2.set_color("blue")
p2.set_cost(521)
p2.play_game()
p2.set_color("red")
p2.set_cost(800)
p2.show_color()
p2.show_cost()
```

```python
# ------------------- CREATING CLASS WITH CONSTRUCTORS------------
class Employee:
    def __init__(self, name,age, salary, gender):
        self.name = name
        self.age = age
        self.salary = salary
        self.gender = gender
    def employee_details(self):
        print("Name of the employee is ",self.name)
        print("Age of ", self.name, "is ", self.age)
        print("Salary of ",self.name, "is ", self.salary)
        print("Gender of ", self.name,  "is ",self.gender)
```

```python
e1 = Employee("Karina",25,11200,"Female")
e1.employee_details()
```
