```
#Lambda function with filter
l1 = [21,34,55,232,2,3,55,71,42]
even_list = list (filter(lambda x:x%2==0,l1))
print("Even numbers out of list are",even_list)
```

```
odd_list = list(filter(lambda x:x%2!=0,l1))
print("Odd numbers out of list are",odd_list)
```

```
#lambda function with map
m_list = list(map(lambda x:x+1,l1))
print(m_list)
```

```
#Lambda function with reduce
from functools import reduce
l2 = [4,9,0,12]
r_list = reduce(lambda x,y:x+y,l2)#here reduce function takes 2 arguments- lambda function and the list. here x,y represents the elements of list l2, x+y will take 4+9, then (4+9= 13,0... then 13+0=13, ... 13+12 and so on)
print(r_list)
```
